# Plot Function

```python
import matplotlib.pyplot as plt
def save_plot(x_input,n=4):
    plt.figure(figsize=(10,7))
    for i in range(n*n):
        plt.subplot(n, n, i+1)
        plt.imshow(x_input[i,:,:,:])
        plt.axis('off')
    plt.savefig("Generated Images.png")
    plt.show()
```

# Load Generator Model

```python
# load generator model
generator = tensorflow.keras.models.load_model("generator_model_epoch_600.h5")
```

# Generate Data

```python
def generate_latent_points(latent_dim, n_samples):
    x_input = randn(latent_dim * n_samples)
    z_input = x_input.reshape(n_samples, latent_dim)
    z = np.random.randint(0,5,size=(n_samples))
    return [z_input,z]
```

```python
from numpy import *
from numpy.random import *

# generated images
latent_dim = 512
n_samples = 100
latent_points, labels = generate_latent_points(latent_dim= latent_dim, n_samples= n_samples)
pred  = generator.predict([latent_points, labels])
```

# Plot Generated Data

```python
# Generated images
X1 = pred[:100]
X1 = (X1 + 1)/2.0
save_plot(X1,n=6)
```

# Checking Variation

# 1:- Different Label

```python
def save_plot(x_input,n=4):
    plt.figure(figsize=(10,7))
    for i in range(n*n):
        plt.subplot(n, n, i+1)
        plt.imshow(x_input[i])
        plt.axis('off')
    plt.show()
    
def generate_latent_points(latent_dim, n_samples):
    x_input = randn(latent_dim * n_samples)
    z_input = x_input.reshape(n_samples, latent_dim)
    z = np.random.randint(0,5,size=(n_samples))
    return [z_input,z]

# generated images
latent_dim = 512
n_samples = 100
latent_points, labels = generate_latent_points(latent_dim= latent_dim, n_samples= n_samples)
print("Label: ", labels)
pred  = generator.predict([latent_points, labels], verbose=0)

# Generated images
X1 = pred[:100]
X1 = (X1 + 1)/2.0
save_plot(X1,n=6)
```
