```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import load_img
```

```python
path = "DELAUNAY_test"
# print(os.listdir(path)[:5])
```

```python
image_list=[]
for i in os.listdir(path):
    for j in os.listdir(str(path)+"/"+str(i)):
        path_image = str(path)+"/"+str(i)+"/"+str(j)
        image = load_img(path_image,target_size=(128,128))
        image = np.array(image)
        image_list.append(image)
```

```python
np.savez_compressed("Abstract_128_combo.npz", a=image_list)
```
