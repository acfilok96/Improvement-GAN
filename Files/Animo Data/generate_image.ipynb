```python
def save_plot(x_input,n=4):
    plt.figure(figsize=(7,5))
    for i in range(n*n):
        plt.subplot(n, n, i+1)
        plt.imshow(x_input[i,:,:,:])
        plt.axis('off')
    plt.show()
# plot data
# save_plot(k[:16])
```

```python
def generate_latent_points(latent_dim, n_samples):
    x_input = randn(latent_dim * n_samples)
    z_input = x_input.reshape(n_samples, latent_dim)
    z = np.random.randint(0,7,size=(n_samples))
    return [z_input,z]
# p = generate_latent_points(512, 32)
# print("Generate latent point(with label) as a batch: ",p[0].shape)
```

```python
generator = tensorflow.keras.models.load_model("generator_model.h5")
```

```python
latent_dim = 512
n_samples = 16
z_input, labels = generate_latent_points(latent_dim, n_samples)
data = [z_input,labels]
pred = generator.predict(data,verbose=0)
pred = (pred + 1 ) / 2.0
save_plot(pred,n=4)
```
